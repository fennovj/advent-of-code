```python
import advent

def map_f(line):
    from_, to = line.split(": ")
    to = to.split(" ")
    return from_, to

data = dict(advent.get_lines(11, map_fn=map_f))
```

```python
def count_paths(node):
    if node == "out": return 1
    return sum(count_paths(neighbor) for neighbor in data[node])

print(count_paths("you"))
```

```python
from functools import cache

@cache
def count_paths_2(node, visited_fft=False, visited_dac=False):
    if node == 'out':
        return visited_dac and visited_fft
    total = 0
    for neighbor in data[node]:
        total += count_paths_2(
            neighbor,
            visited_fft or neighbor == 'fft',
            visited_dac or neighbor == 'dac'
        )
    return total

print(count_paths_2("svr"))
```
