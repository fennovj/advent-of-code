```python
import advent
data = advent.get_lines(1)
```

```python
def part1(line, prev, modulo=100):
    c, i = line[0], line[1:]
    if c == 'R':
        return (prev + int(i)) % modulo
    elif c == 'L':
        return (prev - int(i)) % modulo
    raise ValueError(f'Unknown direction: {c}')

value, count = 50, 0
for line in data:
    value = part1(line, value)
    if value == 0: count += 1

print(count)
```

```python
def calc_count(prev, move):
    count = abs(int(move / 100))  # First of all, any full rotations will add 1 count no matter what
    move = move - (int(move / 100) * 100)  # Remove full rotations
    if prev == 0: return count # if we start on 0 we cannot cross it again
    if move > 0 and (prev + move) >= 100: count += 1
    if move < 0 and (prev + move) <= 0: count += 1
    return count

def part2(line, prev):
    c, i = line[0], line[1:]
    if c == 'R':
        new_value = (prev + int(i)) % 100
        count = calc_count(prev, int(i))
    elif c == 'L':
        new_value =  (prev - int(i)) % 100
        count = calc_count(prev, -int(i))
    return new_value, count

value, count = 50, 0
for line in data:
    value, count_ = part2(line, value)
    count += count_

print(count)
```
