```python
import advent
from scipy.ndimage import generic_filter
data_raw = advent.get_char_grid(4)
# data is numpy array with # and ., replace it with 1 and 0
data = (data_raw == '@').astype(int)
```

```python
def apply(values):
    return ((values.sum() - values[4]) < 4) and values[4]

generic_filter(data, apply, size=3, mode='constant', cval=0).sum()
```

```python
# Part 2

def remove(data):
    return data - generic_filter(data, apply, size=3, mode='constant', cval=0)

data = (data_raw == '@').astype(int)
original_rolls = data.sum()

for _ in range(100): # just do 100 iterations because why not :)
    data = remove(data)

print(original_rolls - data.sum())
```
