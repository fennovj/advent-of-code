```python
import advent
import numpy as np
data = advent.get_char_grid(7)
start = np.where(data[0] == 'S')[0][0]
data = data == '^'
```

```python
row, result = 0, 0
beams = set([start])
while row < data.shape[0]:
    new_beams = set([])
    for beam in beams:
        if data[row, beam]:
            new_beams.add(beam - 1)
            new_beams.add(beam + 1)
            result += 1
        else:
            new_beams.add(beam)
    beams = new_beams
    row += 1
print(result)
```

```python
# Well this seems like a job for lazy caching since there are only a few rows/columns
from functools import cache

@cache
def number_of_timelines(row, beam):
    if row == data.shape[0]: return 1
    if data[row, beam]:
        return number_of_timelines(row + 1, beam - 1) + number_of_timelines(row + 1, beam + 1)
    else:
        return number_of_timelines(row + 1, beam)

print(number_of_timelines(0, start))
```
