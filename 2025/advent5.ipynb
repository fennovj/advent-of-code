```python
import advent
data = advent.get_lines_doublenewline(5)
ranges = [list(map(int, line.split('-'))) for line in data[0]]
data = list(map(int, data[1]))
```

```python
count = 0
for number in data:
    for r in ranges:
        if r[0] <= number <= r[1]:
            count += 1
            break
print(count)
```

```python
def merge_ranges(r1, r2):
    if r1[1] < r2[0]: return None
    if r2[1] < r1[0]: return None
    return [min(r1[0], r2[0]), max(r1[1], r2[1])]

def merge_all_ranges(candidate, ranges):
    for r in ranges:
        merged = merge_ranges(candidate, r)
        # If we can merge, try to merge it again with the rest
        # e.g. if we merge 12-18 into [10-13, 17-20], we first merge to [10-18, 17-20], but then need to re-merge to 10-20
        if merged is not None:
            return merge_all_ranges(merged, [x for x in ranges if x != r])
    return ranges + [candidate]

# Start with empty ranges, merge all ranges one by one
result = []
for r in ranges:
    result = merge_all_ranges(r, result)

total = 0
for r in result:
    total += r[1] - r[0] + 1
print(total)
```
