```python
import advent
data = advent.get_lines(3)
```

```python
def find_largest(line):
    largest = sorted(line[:-1])[-1]
    ix0 = line.index(largest)
    rest = line[ix0+1:]
    rest_largest = sorted(rest)[-1]
    return int(largest + rest_largest)

print(sum(find_largest(line) for line in data))
```

```python
def find_largest_n(line, n):
    assert len(line) >= n
    if n == 1: return int(sorted(line)[-1])
    largest = sorted(line[:-(n-1)])[-1]
    ix0 = line.index(largest[0])
    rest = line[ix0+1:]
    return int(largest + str(find_largest_n(rest, n-1)))


print(sum(find_largest_n(line, 12) for line in data))
```
