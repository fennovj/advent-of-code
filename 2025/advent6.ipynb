```python
import advent
import math
data = advent.get_lines(6)
operators = data[4].split()
data = [list(map(int, data[i].split())) for i in range(4)]
```

```python
result = 0
for i, op in enumerate(operators):
    if op == '+':
        result += sum([data[j][i] for j in range(4)])
    elif op == '*':
        result += math.prod([data[j][i] for j in range(4)])

print(result)
```

```python
# part 2
import numpy as np
data = advent.get_char_grid(6, 'txt')

result = 0
op_ixs = np.where(data[4] != ' ')[0]
for ix, op_ix in enumerate(op_ixs):
    # Get beginning/end columns of this particular operation
    op_end = op_ixs[ix + 1] - 1 if ix + 1 < len(op_ixs) else data.shape[1]

    op = sum if data[4][op_ix] == '+' else math.prod
    subresult = 0 if data[4][op_ix] == '+' else 1
    for col_ix in range(op_ix, op_end):  # For each column in this particular operation:
        col_values = [data[row_ix][col_ix] for row_ix in range(4)] # Get the column values with numpy index nonsense
        col_values = int(''.join(map(str, col_values)))  # Convert list of digits to integer
        subresult = op([subresult, col_values]) # Either add or multipy the subresult
    result += subresult

print(result)
    
```
