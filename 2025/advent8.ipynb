```python
import advent
from math import pow

tup3 = tuple[int, int, int]

data: list[tup3] = advent.get_lines(8, map_fn = lambda x: [int(c) for c in x.split(',')])

def euclid(a: tup3, b: tup3) -> int:
    return pow(abs(a[0] - b[0]), 2) + pow(abs(a[1] - b[1]), 2) + pow(abs(a[2] - b[2]), 2)


distances = [[(euclid(data[i], data[j]), i, j) for i in range(j)] for j in range(len(data))]
distances = sorted([d for e in distances for d in e]) # flattened and sorted
```

```python
junctions: list[set[tup3]] = []

for i in range(1000):
    # Combine the unchanged junctions with the changed ones which are combined into one big junction
    unchanged = [junc for junc in junctions if distances[i][1] not in junc and distances[i][2] not in junc]
    changed = [junc for junc in junctions if distances[i][1] in junc or distances[i][2] in junc]
    junctions = unchanged + [set.union(*changed, set([distances[i][1], distances[i][2]]))]

biggest_junctions = sorted(junctions, key = lambda x: -len(x))
print(len(biggest_junctions[0]) * len(biggest_junctions[1]) * len(biggest_junctions[2]))
```
