```python
import advent
data = advent.get_lines(2)[0].split(",")
```

```python
def valid_passwords(start, end, divisor=2):
    assert abs(len(start) - len(end)) <= 1 # safety check
    # Both are odd, there cannot be valid passwords
    if len(start) % divisor != 0 and len(end) % divisor != 0: return 0
    # The next logic was updated a bit for part 2 :) but basically the same just that divisor is always 2 for part 1
    elif len(start) % divisor == 0 and len(end) % divisor == 0: pass
    elif len(end) % divisor == 0 and len(start) % divisor == (divisor - 1): start = '1' + '0' * len(start)
    elif len(start) % divisor == 0 and len(end) % divisor == 1: end = '9' * len(start)
    elif len(start) % divisor == 0: return 0
    elif len(end) % divisor == 0: return 0

    # Now both start and end are even lengths so we can just brute force the first half and double it
    shalf, ehalf = start[:len(start)//divisor], end[:len(end)//divisor]
    for candidate in range(int(shalf), int(ehalf) + 1):
        candidate_ = int(str(candidate) * divisor)
        if candidate_ >= int(start) and candidate_ <= int(end):
            RESULTS.add(candidate_)

RESULTS = set() 

for line in data:
    start, end = line.split("-")
    valid_passwords(start, end, 2)

print(sum(RESULTS))
```

```python
RESULTS = set()

for divisor in range(2, 10): # 10 probably bigger than neccesary but who cares
    for line in data:
        start, end = line.split("-")
        valid_passwords(start, end, divisor)
print(sum(RESULTS))
```
