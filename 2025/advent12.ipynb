```python
import advent
data = advent.get_lines_doublenewline(12)
problem = [line.split(" ") for line in data[-1]]

problem = [((int(p[0][0:2]), int(p[0][3:5])), [int(i) for i in p[1:]]) for p in problem]

size_dict = {0: 6, 1: 7, 2: 5, 3: 7, 4: 7, 5: 7}
```

```python
# Simple heuristic: if we can just put each space in its own 3x3 block, it's always possible
# If the total area of the blocks is less than the area of the grid, it's impossible

def simple_heuristic(problem):
    result = []
    for line in problem:
        total_size = line[0][0] * line[0][1]
        total_size_grid_three = (line[0][0] // 3) * (line[0][1] // 3)
        total_pieces = sum(line[1])

        minimum_size = sum(size_dict[i] * line[1][i] for i in range(6))
        if minimum_size > total_size: result.append("Impossible")
        elif total_pieces <= total_size_grid_three: result.append("Possible")
        else: result.append("Unknown")
    return result

# Count how many are possible, impossible, unknown
results = simple_heuristic(problem)
possible = results.count("Possible")
impossible = results.count("Impossible")
unknown = results.count("Unknown")
print(f"Possible: {possible}, Impossible: {impossible}, Unknown: {unknown}")

        
```
